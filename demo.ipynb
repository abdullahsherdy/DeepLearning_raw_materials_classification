```python
import os, shutil, random

root = "minc-2500" 
train_dir = "dataset/train"
val_dir = "dataset/val"
test_dir = "dataset/test"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

classes = os.listdir(root)

for cls in classes:
    cls_path = os.path.join(root, cls)
    imgs = os.listdir(cls_path)
    random.shuffle(imgs)

    train_cut = int(0.7 * len(imgs))
    val_cut = int(0.15 * len(imgs))

    train_imgs = imgs[:train_cut]
    val_imgs = imgs[train_cut:train_cut + val_cut]
    test_imgs = imgs[train_cut + val_cut:]

    os.makedirs(f"{train_dir}/{cls}", exist_ok=True)
    os.makedirs(f"{val_dir}/{cls}", exist_ok=True)
    os.makedirs(f"{test_dir}/{cls}", exist_ok=True)

    for img in train_imgs:
        shutil.copy(os.path.join(cls_path, img), f"{train_dir}/{cls}/{img}")

    for img in val_imgs:
        shutil.copy(os.path.join(cls_path, img), f"{val_dir}/{cls}/{img}")

    for img in test_imgs:
        shutil.copy(os.path.join(cls_path, img), f"{test_dir}/{cls}/{img}")

print("Dataset splitting completed.")
```

## Preprocessing Pipeline 
### Image Specs (Standard for VGG19/ResNet/Inception/MobileNet/VIT)

- size = 224 × 224
- channels = 3
- normalization: [0,1] → then model-specific preprocessing

```python
import tensorflow as tf

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/train",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/val",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/test",
    seed=42,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
```
